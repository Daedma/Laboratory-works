3. Создать вектор-строку 1x5 из случайных целых чисел. Вычислить норму ‖x‖5

проверить результат с помощью linalg.norm().


```python
# Создание вектор-строки
x = np.random.randint(0, 10, size=5)

def norm(x, p):
    return np.sum(np.abs(x)**p)**(1/p)

# Вычисление нормы ‖x‖5
norm_custom = norm(x, 5)

# Проверка результата с помощью функции linalg.norm()
norm_np = np.linalg.norm(x, ord=5)

# Вывод результатов на экран
print("Вектор-строка x:")
print(x)
print("Норма ‖x‖5, вычисленная самостоятельно:")
print(norm_custom)
print("Норма ‖x‖5, вычисленная с помощью функции linalg.norm():")
print(norm_np)
```

матрицы ‖A‖1

```python
# Создание матрицы
A = np.random.randint(-10, 10, size=(5, 5))

# Написание функции вычисления нормы
def norm(A, p):
    if p == 1:
        return np.max(np.sum(np.abs(A), axis=0))
    elif p == np.inf:
        return np.max(np.sum(np.abs(A), axis=1))
    elif p == 2:
        return np.sqrt(np.sum(np.square(A)))
    else:
        raise ValueError("Неподдерживаемое значение p")

# Вычисление нормы ‖A‖1
norm_custom = norm(A, 2)

# Проверка результата с помощью функции linalg.norm()
norm_np = np.linalg.norm(A)

# Вывод результатов на экран
print("Матрица A:")
print(A)
print("Норма ‖A‖1, вычисленная самостоятельно:")
print(norm_custom)
print("Норма ‖A‖1, вычисленная с помощью функции linalg.norm():")
print(norm_np)
```

Например, вектор (1,2,3,4,5) переводит в вектор (a,0,0,0,0).


```python
import numpy as np

def householder_reflection(vector):
    """
    Функция для нахождения отражения Хаусхолдера для одномерного массива
    """
    # Длина вектора
    n = len(vector)

    # Вычисляем норму вектора
    norm = np.linalg.norm(vector)

    # Вычисляем знак первого элемента вектора
    sign = 1 if vector[0] >= 0 else -1

    # Создаем новый вектор, в котором первый элемент равен длине вектора со знаком
    v = np.zeros(n)
    v[0] = -sign * norm

    # Вычитаем из нового вектора исходный вектор
    v = v - vector

    # Вычисляем матрицу отражения Хаусхолдера
    H = np.identity(n) - 2 * np.outer(v, v) / np.dot(v, v)

    # Умножаем матрицу отражения на исходный вектор
    result = H.dot(vector)

    return np.round(result, 8)
print(householder_reflection(x))
```

6. Для матрицы, созданной в 4 пункте, найти LU разложение
самостоятельно написанной функцией и проверить результат с
помощью перемножения матриц

```python
def lu_decomposition(mtx):
    n = len(mtx)
    lower = np.zeros((n, n))
    upper = np.zeros((n, n))

    for i in range(n):
        lower[i][i] = 1

        for j in range(i, n):
            Sum = 0
            for k in range(i):
                Sum += (lower[i][k] * upper[k][j])
            upper[i][j] = mtx[i][j] - Sum

        for j in range(i+1, n):
            Sum = 0
            for k in range(i):
                Sum += (lower[j][k] * upper[k][i])
            if upper[i][i] == 0:
                return None
            lower[j][i] = (mtx[j][i] - Sum) / upper[i][i]

    return lower, upper

# Пример использования функции
print("Матрица A:")
print(A)
result = lu_decomposition(A)

if result is None:
    print("Матрица не может быть разложена в LU форму")
else:
    L, U = result
    # Проверка, что LU разложение верно
    print("L = ", L)
    print("U = ", U)
    res = np.dot(L, U)
    res = np.round(res).astype(int) # округление элементов матрицы res до int
    if np.allclose(res, A, rtol=1e-5):
        print("LU разложение верно!")
    else:
        print("LU разложение неверно!")
    print("A = ",res)
```

```python
def householder_reflection(v):
    norm = np.linalg.norm(v)
    if norm == 0:
        v[0] = 1
    else:
        v = v / norm
        if v[0] >= 0:
            v[0] += 1
        else:
            v[0] -= 1
        norm = np.linalg.norm(v)
        v = v / norm
    return v

def qr_householder(A):
    m, n = A.shape
    Q = np.eye(m, dtype=np.float64)
    R = A.copy().astype(np.float64)
    for i in range(n):
        v = R[i:, i].copy()
        v[0] = v[0] + np.sign(v[0]) * np.linalg.norm(v)
        v = householder_reflection(v)
        H = np.eye(m)
        H[i:, i:] -= 2.0 * np.outer(v, v)
        Q = Q.dot(H)
        R = H.dot(R)
    return Q, R
print("Матрица A:\n", A)
Q1, R1 = qr_householder(A)
print("householder")
print("Матрица Q:")
print(Q1)
print("Матрица R:")
print(R1)
print("Q * R:")
print(Q1.dot(R1))
Qb, Rb = np.linalg.qr(A)
def gram_schmidt(A):
    m, n = A.shape
    Q = np.zeros(A.shape, dtype=np.float64)
    R = np.zeros((A.shape[1], A.shape[1]),dtype=np.float64)

    for j in range(n):
        v = A[:, j].astype(np.float64)
        for i in range(j):
            R[i, j] = Q[:, i].T.dot(A[:, j])
            v -= R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]

    return Q, R
Q2, R2 = gram_schmidt(A)
print("gram_schmidt")
print("Матрица Q:")
print(Q2)
print("Матрица R:")
print(R2)
print("Q * R:")
print(Q2.dot(R2))
Qb, Rb = np.linalg.qr(A)
print("np.linalg.qr()")
print("Матрица Q:")
print(Qb)
print("Матрица R:")
print(Rb)
print("Q * R:")
print(Qb.dot(Rb))
```
