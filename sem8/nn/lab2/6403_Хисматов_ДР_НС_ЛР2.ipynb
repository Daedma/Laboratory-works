2. Попробуем обучить сеть, изменяя вес в разные стороны и рассчитывая значение ошибки для каждого изменения

```python
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001


for iteration in range(10):

    prediction = input * weight
    error = (prediction - goal_prediction) ** 2

    print("Error:" + str(error) + " Prediction:" + str(prediction))

    up_prediction = input * (weight + step_amount) #попробовать увеличить
    up_error = (goal_prediction - up_prediction) ** 2

    down_prediction = input * (weight - step_amount) #попробовать уменьшить
    down_error = (goal_prediction - down_prediction) ** 2

    if(down_error < up_error):
        weight = weight - step_amount #если уменьшение дало лучший результат, уменьшить

    if(down_error > up_error):
        weight = weight + step_amount # если увеличение дало лучший результат, увеличить
```

3. Реализуйте градиентный спуск: определите направление и величину изменения веса, используя производную функции потерь для текущего веса . Выполните 20 итераций градиентного спуска .
 weight = weight - derivative

```python
# Инициализация начального веса
weight = 0.5

# Целевое значение, которое мы хотим предсказать
goal_prediction = 0.8

# Входное значение для модели
input = 0.5

# Функция потерь, которая измеряет, насколько наше предсказание отличается от целевого значения
def loss(w):
  return (goal_prediction - w * input) ** 2  # Вычисляем квадрат разницы между целевым значением и предсказанием

# Функция для вычисления градиента функции f в точке x с шагом dx
def grad(f, x, dx=0.0001):
  return (f(x + dx / 2) - f(x - dx / 2)) / dx  # Используем численное дифференцирование для вычисления градиента

# Основной цикл обучения модели
for iteration in range(20):  # Повторяем 20 итераций
  prediction = weight * input  # Вычисляем текущее предсказание модели
  error = loss(weight)  # Вычисляем текущую ошибку (значение функции потерь)
  derivative = grad(loss, weight)  # Вычисляем градиент функции потерь по весу
  weight = weight - derivative  # Обновляем вес, уменьшая его на значение градиента (градиентный спуск)
  print("Error:" + str(error) + " Prediction:" + str(prediction))  # Выводим текущую ошибку и предсказание

```

4. Реализуйте алгоритм градиентного спуска, добавив параметр скорости обучения.

```python
# Инициализация начального веса
weight = 0.5

# Целевое значение, которое мы хотим предсказать
goal_prediction = 0.8

# Входное значение для модели
input = 0.5

# Скорость обучения, которая определяет, насколько сильно мы корректируем вес на каждой итерации
learning_rate = 2.0

# Функция потерь, которая измеряет, насколько наше предсказание отличается от целевого значения
def loss(w):
  return (goal_prediction - w * input) ** 2  # Вычисляем квадрат разницы между целевым значением и предсказанием

# Функция для вычисления градиента функции f в точке x с шагом dx
def grad(f, x, dx=0.0001):
  return (f(x + dx / 2) - f(x - dx / 2)) / dx  # Используем численное дифференцирование для вычисления градиента

# Основной цикл обучения модели
for iteration in range(20):  # Повторяем 20 итераций
  prediction = weight * input  # Вычисляем текущее предсказание модели
  error = loss(weight)  # Вычисляем текущую ошибку (значение функции потерь)
  derivative = grad(loss, weight)  # Вычисляем градиент функции потерь по весу
  weight = weight - derivative * learning_rate  # Обновляем вес, уменьшая его на произведение градиента и скорости обучения
  print("Error:" + str(error) + " Prediction:" + str(prediction))  # Выводим текущую ошибку и предсказание
```
